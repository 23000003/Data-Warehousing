### Assign Unique Column to reference to remove duplicates in Dimension

#### Dimension Generator


```python
# Create a new column 'Incident_Unique_ID' to store the reference values
df_crashes['Incident_Unique_ID'] = -1

# Group by the specified columns and assign a unique ID to each group
df_crashes['Incident_Unique_ID'] = df_crashes.groupby(['REPORT_TYPE', 'INTERSECTION_RELATED_I', 'HIT_AND_RUN_I', 'NUM_UNITS']).ngroup()

# Fill NaN values with a specific integer value before converting to integer type
df_crashes['Incident_Unique_ID'] = df_crashes['Incident_Unique_ID'].fillna(-1).astype(int)

# Verify the changes
print(df_crashes[['REPORT_TYPE', 'INTERSECTION_RELATED_I', 'HIT_AND_RUN_I', 'NUM_UNITS', 'Incident_Unique_ID']].head())
```
