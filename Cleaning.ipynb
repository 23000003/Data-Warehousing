```python
import pandas as pd
```

```python
# Read the CSV file from the 'new' folder
df_crashes = pd.read_csv('../Traffic_Crashes_-_Crashes.csv')

# Display the first few rows of the dataframe
df_crashes.head()
```

```python


iRows, iColumns = df_crashes.shape


print(f"Total rows: {iRows}")
print(f"Total columns: {iColumns}")
```

```python
# Drop rows where both LONGITUDE and LATITUDE are NaN
df_crashes.dropna(subset=['LONGITUDE', 'LATITUDE'], how='all', inplace=True)

# Display the first few rows of the dataframe to confirm the changes
df_crashes.head()
```

```python
iRows, iColumns = df_crashes.shape


print(f"Total rows: {iRows}")
print(f"Total columns: {iColumns}")
```

```python
# Convert the CRASH_DATE column to datetime format
df_crashes['CRASH_DATE'] = pd.to_datetime(df_crashes['CRASH_DATE'])

# Extract the year from the CRASH_DATE column
df_crashes['CRASH_YEAR'] = df_crashes['CRASH_DATE'].dt.year
df_crashes['CRASH_DAY'] = df_crashes['CRASH_DATE'].dt.day
df_crashes['CRASH_TIME'] = df_crashes['CRASH_DATE'].dt.time
df_crashes['CRASH_MERIDIAN'] = df_crashes['CRASH_DATE'].dt.strftime('%p')

# Group by the year and count the number of rows per year
crashes_per_year = df_crashes.groupby('CRASH_YEAR').size()

# Print the count of number of rows per year
print(crashes_per_year)
```

```python
# Drop rows where CRASH_YEAR is 2018 or below
df_crashes = df_crashes[df_crashes['CRASH_YEAR'] >= 2015]

# Display the first few rows of the dataframe to confirm the changes
df_crashes.head()
```

```python
iRows, iColumns = df_crashes.shape

print(f"Total rows: {iRows}")
print(f"Total columns: {iColumns}")
```

```python
# Group by the year and count the number of rows per year
crashes_per_year = df_crashes.groupby('CRASH_YEAR').size()

# Print the count of number of rows per year

print(crashes_per_year)
```

```python
# Convert the DATE_POLICE_NOTIFIED column to datetime format
df_crashes['DATE_POLICE_NOTIFIED'] = pd.to_datetime(df_crashes['DATE_POLICE_NOTIFIED'], errors='coerce')

# Extract the year from the DATE_POLICE_NOTIFIED column
df_crashes['DATE_POLICE_NOTIFIED_YEAR'] = df_crashes['DATE_POLICE_NOTIFIED'].dt.year

# Display the first few rows of the dataframe to confirm the changes
df_crashes.head()
# Extract the day from the DATE_POLICE_NOTIFIED column
df_crashes['DATE_POLICE_NOTIFIED_DAY'] = df_crashes['DATE_POLICE_NOTIFIED'].dt.day

# Extract the month from the DATE_POLICE_NOTIFIED column
df_crashes['DATE_POLICE_NOTIFIED_MONTH'] = df_crashes['DATE_POLICE_NOTIFIED'].dt.month

# Extract the time from the DATE_POLICE_NOTIFIED column
df_crashes['DATE_POLICE_NOTIFIED_TIME'] = df_crashes['DATE_POLICE_NOTIFIED'].dt.strftime('%I:%M:%S')

# Extract the meridian (AM/PM) from the CRASH_DATE column
df_crashes['DATE_POLICE_NOTIFIED_MERIDIAN'] = df_crashes['DATE_POLICE_NOTIFIED'].dt.strftime('%p')

# Display the first few rows of the dataframe to confirm the changes
df_crashes.head()
```

```python
columns_to_keep = [
    'CRASH_DATE', 'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE', 'ROADWAY_SURFACE_COND', 
    'REPORT_TYPE', 'INTERSECTION_RELATED_I', 'HIT_AND_RUN_I', 'DAMAGE', 'DATE_POLICE_NOTIFIED', 
    'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE', 'STREET_NO', 'STREET_DIRECTION', 'STREET_NAME', 
    'NUM_UNITS', 'INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING', 
    'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN', 'CRASH_MONTH', 
    'LATITUDE', 'LONGITUDE', 'CRASH_YEAR', 'CRASH_DAY', 'CRASH_TIME', 'CRASH_MERIDIAN', 
    'DATE_POLICE_NOTIFIED_TIME', 'DATE_POLICE_NOTIFIED_YEAR', 'DATE_POLICE_NOTIFIED_DAY', 
    'DATE_POLICE_NOTIFIED_MONTH', 'DATE_POLICE_NOTIFIED_MERIDIAN'
]

df_crashes = df_crashes[columns_to_keep]

# Display the first few rows of the dataframe to confirm the changes
df_crashes.head()
```
